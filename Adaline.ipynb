```python
def entrenar_red(Xp, YDp, alfa, precision, max_epocas=10000):
    num_patrones, num_entradas = Xp.shape
    W = inicializar_pesos(num_entradas)
    epocas = inicializar_epocas()
    error_anterior = 0
    error_actual = float('inf')

    errores = []  # Lista para guardar el error de cada época

    while error_estable(error_actual, error_anterior) > precision and epocas < max_epocas:
        error_anterior = error_actual
        error_acumulado = 0
     # Paso 6.2: Calcular la salida ponderada Yo para cada patrón
        for i in range(num_patrones):
            Xp_i = Xp[i]
            YDp_i = YDp[i]
            Yo = np.dot(W, Xp_i)  # Salida ponderada Yo = Σ Wi * Xi

            # Paso 6.3: Actualizar los pesos
            W += alfa * (YDp_i - Yo) * Xp_i  # Regla de aprendizaje delta

            # Paso 6.4: Calcular el error cuadrático
            error_acumulado += (YDp_i - Yo) ** 2
            error_actual = error_acumulado / num_patrones
            epocas += 1
            errores.append(error_actual)  # Guardar el error de esta época
            print(f"Época {epocas}, Error actual: {error_actual}")







    plt.plot(errores)
    plt.title('Evolución del error durante el entrenamiento')
    plt.xlabel('Épocas')
    plt.ylabel('Error')
    plt.grid(True)
    plt.show()

    return W, epocas, error_actual

# Guardar el resultado de la red en un archivo .txt
def guardar_resultado(W, epocas, error_actual, archivo_salida):
    with open(archivo_salida, 'w') as archivo:
        archivo.write(f"Pesos finales: {W}\n")
        archivo.write(f"Épocas totales: {epocas}\n")
        archivo.write(f"Error final: {error_actual}\n")

# Función para subir archivos desde el dispositivo
def cargar_archivos():
    uploaded = files.upload()  # Permite subir el archivo desde el dispositivo
    nombre_archivo = list(uploaded.keys())[0]
    print(f"Archivo '{nombre_archivo}' cargado con éxito.")
    return nombre_archivo

# Función principal para la ejecución
def main(alfa, precision):
    # Cargar los archivos .txt desde el dispositivo
    print("Por favor, sube el archivo de patrones de entrada (Xp):")
    archivo_entradas = cargar_archivos()

    print("Por favor, sube el archivo de salidas deseadas (YDp):")
    archivo_salidas = cargar_archivos()

    # Cargar los datos de entrada (Xp) y las salidas deseadas (YDp)
    Xp, YDp = cargar_datos(archivo_entradas, archivo_salidas)

    # Entrenar la red neuronal
    W, epocas, error_actual = entrenar_red(Xp, YDp, alfa, precision)

    # Guardar los resultados en un archivo .txt
    archivo_salida = '/content/resultados.txt'
    guardar_resultado(W, epocas, error_actual, archivo_salida)

    print(f"Entrenamiento completado en {epocas} épocas.")
    print(f"Pesos finales guardados en: {archivo_salida}")

# Crear los cuadros de texto dinámicos para alfa y precisión
alfa_text = widgets.FloatText(
    value=0.01,
    description='Alfa:',
    disabled=False
)

precision_text = widgets.FloatText(
    value=0.0001,
    description='Precisión:',
    disabled=False
)

# Botón para iniciar el entrenamiento
button = widgets.Button(description="Iniciar entrenamiento")


def on_button_click(b):
    alfa = alfa_text.value
    precision = precision_text.value
    main(alfa, precision)


button.on_click(on_button_click)


display(alfa_text)
display(precision_text)
display(button)
```
